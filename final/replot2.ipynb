```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()


def plot_metrics(
    data_list,
    label_list,
    colour_list,
    plot_name,
    AUC=True,
):
    min_epochs = min(len(data['Train Loss']) for data in data_list)
    plt.figure(figsize=(16, 7))
    plt.subplot(1, 2, 1)
    for idx, data in enumerate(data_list):
        epochs = list(range(1, min_epochs + 1))
        plt.plot(epochs, data['Train Loss'][:min_epochs], label=f'{label_list[idx]} Train', color=colour_list[idx], linestyle='solid')
        plt.plot(epochs, data['Val Loss'][:min_epochs], label=f'{label_list[idx]} Val', color=colour_list[idx], linestyle='dashed')
    plt.title('Train and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    # plt.legend()
    # plt.tight_layout()
    # plt.savefig(f"pdf3/{plot_name}_loss.pdf", format="pdf", bbox_inches='tight', facecolor='w')
    # plt.show()
    
    plt.subplot(1, 2, 2)
    for idx, data in enumerate(data_list):
        epochs = list(range(1, min_epochs + 1))
        plt.plot(epochs, data['Train Acc'][:min_epochs], label=f'{label_list[idx]} Train', color=colour_list[idx], linestyle='solid')
        plt.plot(epochs, data['Val Acc'][:min_epochs], label=f'{label_list[idx]} Val', color=colour_list[idx], linestyle='dashed')
    plt.title("Train and Validation Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    # plt.legend()
    plt.legend(loc='center right', bbox_to_anchor=(1, 0.575))
    plt.tight_layout()
    plt.savefig(f"pdf4/{plot_name}.pdf", format="pdf", bbox_inches='tight', facecolor='w')
    plt.show()
  

    # Bottom row: Train and Val AUC
    # if AUC:
    #     plt.figure(figsize=(12, 7))
    #     for idx, data in enumerate(data_list):
    #         epochs = list(range(1, min_epochs + 1))
    #         plt.plot(epochs, data['Train AUC'][:min_epochs], label=f'{label_list[idx]} Train', color=colour_list[idx], linestyle='solid')
    #         plt.plot(epochs, data['Val AUC'][:min_epochs], label=f'{label_list[idx]} Val', color=colour_list[idx], linestyle='dashed')
    #     plt.title("Train and Validation AUC")
    #     plt.xlabel("Epochs")
    #     plt.ylabel("AUC")
    #     plt.legend()
    #     plt.tight_layout()
    #     plt.savefig(f"pdf2/{plot_name}_auc.pdf", format="pdf", bbox_inches='tight', facecolor='w')
    #     plt.show()
```
