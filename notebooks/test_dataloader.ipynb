```python
import re


def extract_batch_number(file_path):
    match = re.search(r"_batch_(\d+).pt", file_path)
    if match:
        return int(match.group(1))
    else:
        return -1  # Or raise an error if the pattern doesn't match
```

```python
group_dir = os.getenv("MYGROUP")
file_list = glob.glob(
    f"{group_dir}/QuantumTransformer/data/word_embeddings/{name}_{type}_batch_*.pt"
)
sorted(file_list, key=extract_batch_number)
```
